```python
import pandas as pd
import json

# Read the CSV file into a pandas dataframe
df = pd.read_csv('/Users/bjxl/fyp/csv/user_review_count_more_than_2000.csv')

# Define the compliment columns to keep
compliment_cols = ['compliment_hot', 'compliment_note',
                   'compliment_plain', 'compliment_cool', 'compliment_funny', 'compliment_photos']

# Loop through each user and generate a JSON file with their compliment data
for user_id, user_data in df.groupby('name'):
    compliments = user_data[compliment_cols].sum()
    compliments = compliments.reset_index()
    compliments.columns = ['name', 'value']
    compliments = compliments.to_dict(orient='records')

    with open(f'/Users/bjxl/fyp/csv/radar/{user_id}.json', 'w') as f:
        json.dump(compliments, f)
```
