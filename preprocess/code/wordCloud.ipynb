```python
user_review_2000=[]
for i in range(0,len(review)):
    for j in user["user_id"]:
        if review["user_id"][i]==j:
            user_review_2000.append([review["review_id"][i],review["user_id"][i],review["text"][i]])
print(user_review_2000)
```

```python
user_review_2000=pd.DataFrame(user_review_2000)
user_review_2000["review_id"]=user_review_2000[0]
user_review_2000["user_id"]=user_review_2000[1]
user_review_2000["text"]=user_review_2000[2]
user_review_2000=user_review_2000[["review_id","user_id","text"]]
user_review_2000
```

```python
df = pd.DataFrame(np.array(user_review_2000["text"]), columns=['review'])
df
```

```python
# Calculate word count
df['word_count'] = df['review'].apply(lambda x: len(str(x).split(" ")))

# Calculate character count
df['char_count'] = df['review'].str.len()

def avg_word(review):
    words = review.split()
    return (sum(len(word) for word in words) / len(words))

# Calculate average words
df['avg_word'] = df['review'].apply(lambda x: avg_word(x))

# Import stopwords
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')

stop_words = stopwords.words('english')
df['stopword_coun'] = df['review'].apply(lambda x: len([x for x in x.split() if x in stop_words]))
```

```python
# Lower case all words
df['review_lower'] = df['review'].apply(lambda x: " ".join(x.lower() for x in x.split()))

# Remove Punctuation
df['review_nopunc'] = df['review_lower'].str.replace('[^\w\s]', '')

# Import stopwords
from nltk.corpus import stopwords
stop_words = stopwords.words('english')

# Remove Stopwords
df['review_nopunc_nostop'] = df['review_nopunc'].apply(lambda x: " ".join(x for x in x.split() if x not in stop_words))
```

```python
# Return frequency of values
freq= pd.Series(" ".join(df['review_nopunc_nostop']).split()).value_counts()[:30]
freq
```

```python
other_stopwords = ['get', 'us', 'see', 'use', 'said', 'asked', 'day', 'go' \
  'even', 'ive', 'right', 'left', 'always', 'would', 'told', \
  'get', 'us', 'would', 'get', 'one', 'ive', 'go', 'even', \
  'also', 'ever', 'x', 'take', 'let' ]
```

```python
df['review_nopunc_nostop_nocommon'] = df['review_nopunc_nostop'].apply(lambda x: "".join(" ".join(x for x in x.split() if x not in other_stopwords)))
```

```python
df.head()
```

```python
#Import textblob
from textblob import Word
import nltk
nltk.download('omw-1.4')
nltk.download('wordnet')
# Lemmatize final review format
df['cleaned_review'] = df['review_nopunc_nostop_nocommon']\
.apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
```

```python
print("Base Review\n",df["review"][0])
print("\n------------------\n")
print("Clean and Lemmatized Review\n",df["cleaned_review"][0])
```

```python
# Define a function to count the frequency of each word
def count_words(data):
    words = data.str.split(expand=True).values.flatten()
    return pd.Series(words).value_counts()

# Group by User and apply the count_words function to get the frequency of each word
word_counts = df.groupby('user_id')['review_nopunc_nostop_nocommon'].apply(count_words)

# Reset index to turn multi-level series into a dataframe
word_counts = word_counts.reset_index(name='count')

# Group again by User and take the top 30 words for each user
top_30_words = word_counts.groupby('user_id').apply(lambda x: x.nlargest(30, 'count'))

# Print the resulting dataframe
top_30_words
```

```python
# Read in the original CSV file
df = pd.read_csv('/Users/bjxl/fyp/csv/user_review_290.csv')

# Get the unique values in the "name" column
unique_names = df['name'].unique()

# Create a dictionary to hold each dataframe for each unique name
dfs_by_name = {}

# Loop through each unique name and create a new dataframe for each one
for name in unique_names:
    dfs_by_name[name] = df.loc[df['name'] == name]

# Loop through the dictionary and write each dataframe to a new CSV file
for name, df_by_name in dfs_by_name.items():
    filename = '/Users/bjxl/fyp/csv/wordCloud/'+name + '.csv'
    df_by_name.to_csv(filename, index=False)
```
